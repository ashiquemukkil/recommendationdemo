```python
data = pd.read_csv("inventory_data.csv")
```

```python
data = data.fillna("none")
```

```python
cleaned_data = data[(data["description"] != "none")&(data["season"] != "kg")]
```

```python
import os
import pandas as pd
import requests

# Assuming your DataFrame is named df
# Make sure to replace 'your_folder_path' with the actual path where you want to save the images.

folder_path = 'images'

# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

def download_image(url, barcode):
    response = requests.get(url)
    if response.status_code == 200:
        with open(os.path.join(folder_path, f'{barcode}.jpg'), 'wb') as file:
            file.write(response.content)
        print(f"Image for barcode {barcode} downloaded successfully.")
    else:
        print(f"Failed to download image for barcode {barcode}. Status code: {response.status_code}")

# Assuming your DataFrame has columns 'variant_code' and 'image_url'
for index, row in cleaned_data.iterrows():
    variant_code = row['variant_code']
    image_url = row['image_url']
    
    download_image(image_url, variant_code)
```
